```python
# Web Scraping of top 100 movies by genre from Rotten tomatoes.
# Just select the genre and get the list of top 100 movies from that genre.

# Importing required libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from urllib.parse import urljoin


# Creating HTTP request for page
base_url='https://www.rottentomatoes.com/top/bestofrt/'
header= {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0'}
# Using header to bypass theuser-agent error for directly non-scrappable sites
response=requests.get(base_url, headers=header)
# Getting content of the page
res_content=response.content

# Parsing the content of the page using lxml parser
soup=bs(res_content,'lxml') 

# Creating function for retreving required list of genre
def list_getter(relative_url):
    new_url= urljoin(base_url, relative_url)
    res= requests.get(new_url)
    
    if res.status_code == 200:
        soup_url= bs(res.content, 'lxml')
        top=pd.read_html(new_url, attrs={'class':'table'})[0]
        top['Rank']=top['Rank'].astype(int)
        top.set_index('Rank', inplace=True)
        return top
    else:
        return 'Error'

pd.set_option('display.max_rows',100)
pd.set_option('display.max_colwidth', None)

# Below is the list from which user can choose to get the required movie list. User is required to select the
# corresponding number to the list
# 1.  Top 100 Action & Adventure Movies
# 2.  Top 100 Animation Movies
# 3.  Top 100 Art House & International Movies
# 4.  Top 100 Classics Movies
# 5.  Top 100 Comedy Movies
# 6.  Top 100 Documentary Movies
# 7.  Top 100 Drama Movies
# 8.  Top 100 Horror Movies
# 9.  Top 100 Kids & Family Movies
# 10. Top 100 Musical & Performing Arts Movies
# 11. Top 100 Mystery & Suspense Movies
# 12. Top 100 Romance Movies
# 13. Top 100 Science Fiction & Fantasy Movies
# 14. Top 100 Special Interest Movies
# 15. Top 100 Sports & Fitness Movies
# 16. Top 100 Television Movies
# 17. Top 100 Western Movies
# 18. Top 100 Movies in all genre

list=int(input("Enter the Number corresponding to list above to get Movie list :"))

# Loop to get the list of movies.
for i in range(1,18):
    if list == 18:
        df=list_getter(base_url)
        break
    elif list >= 19:
        print('Invalid Input ! Try again !!')
        break
    elif list == i:
        div_ul=soup.find('div',{'class':'btn-group btn-primary-border-dropdown'})
        df=list_getter(div_ul.find_all('a')[i-1].get('href'))
        break

df
```
